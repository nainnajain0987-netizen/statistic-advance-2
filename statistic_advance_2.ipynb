#Theory Question

1. What is hypothesis testing in statistics?
   - Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It involves formulating two competing hypotheses (null and alternative), calculating a test statistic from the sample, and determining whether the observed data provides enough evidence to reject the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship in the population. It represents the default assumption.

The alternative hypothesis (H₁ or Ha) is the statement that contradicts the null hypothesis. It suggests that there is an effect, difference, or relationship.

Example:

H₀: The mean weight of products is 50 kg.

H₁: The mean weight of products is not 50 kg.

3.  Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
    - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It represents the risk of making a Type I error.

Common values are 0.05 (5%) or 0.01 (1%).

If the p-value ≤ α, we reject the null hypothesis.
If the p-value > α, we fail to reject the null hypothesis.

Thus, α determines the threshold for deciding whether the results are statistically significant.

4.  What are Type I and Type II errors? Give examples of each.
    - Type I Error: Rejecting a true null hypothesis.

Example: Concluding a medicine works when it actually does not.

Probability = α (significance level).

Type II Error: Failing to reject a false null hypothesis.

Example: Concluding a medicine does not work when it actually does.

Probability = β.

6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000

# Generate binomial data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=11, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50
sigma = np.std(sample_data, ddof=1)
n = len(sample_data)

sample_mean = np.mean(sample_data)

z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z)
print("P-value:", p_value)
```

8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate normal data
data = np.random.normal(loc=50, scale=5, size=100)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% CI
margin_error = stats.t.ppf(0.975, df=n-1) * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print("Mean:", mean)
print("95% Confidence Interval:", (lower, upper))

# Plot
plt.hist(data, bins=15, edgecolor='black')
plt.title("Normal Distribution Data")
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 100)
z_scores = calculate_z_scores(data)

print("First 10 Z-scores:", z_scores[:10])

# Plot
plt.hist(z_scores, bins=15, edgecolor='black')
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
